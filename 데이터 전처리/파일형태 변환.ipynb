```python
import os
import json

# JSON 파일들이 저장된 루트 디렉토리 경로
root_dir = '/content/drive/MyDrive/코딩공부/이것저것 실습폴더/json_files'
# 병합된 JSONL 파일의 저장 위치
output_file = '/content/drive/MyDrive/코딩공부/이것저것 실습폴더/json_merged/merged.jsonl'

# JSONL 파일을 생성하거나 열기 (쓰기 모드)
with open(output_file, 'w', encoding='utf-8') as outfile:
    # os.walk를 사용하여 모든 디렉토리와 파일 탐색
    for dirpath, dirnames, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.endswith('.json'):
                file_path = os.path.join(dirpath, filename)
                with open(file_path, 'r', encoding='utf-8') as infile:
                    data = json.load(infile)

                    # 새로운 형식으로 변환
                    output_json = {
                        "messages": [
                            {
                                "role": "system",
                                "content": "면접관 입장에서 지원자가 가장 해줬으면 하는 답변을 알려주는 역할"
                            },
                            {
                                "role": "user",
                                "content": data["dataSet"]["question"]["raw"]["text"]
                            },
                            {
                                "role": "assistant",
                                "content": data["dataSet"]["answer"]["raw"]["text"]
                            }
                        ]
                    }
                    #print(output_json)
                    # 변환된 데이터를 jsonl 파일에 기록
                    json.dump(output_json, outfile, ensure_ascii=False)
                    outfile.write('\n')  # 각 JSON 객체를 새로운 줄로 분리
        print(dirpath)
print("종료")
```
